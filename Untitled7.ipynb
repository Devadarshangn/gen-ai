```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense
from tensorflow.keras.optimizers import Adam
```

```python
# Load data
data = pd.read_csv('/content/drive/MyDrive/micro dataset/Microsoft_Stock.csv')

# Use 'Close' prices as the target variable
prices = data['Close'].values.reshape(-1, 1)

# Normalize the data
scaler = MinMaxScaler(feature_range=(0, 1))
prices_scaled = scaler.fit_transform(prices)

# Split data into training and testing sets
train_size = int(len(prices_scaled) * 0.8)
train_data = prices_scaled[:train_size]
test_data = prices_scaled[train_size:]
```

```python
def create_sequences(data, seq_length):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i + seq_length])
        y.append(data[i + seq_length])
    return np.array(X), np.array(y)

# Define sequence length (number of previous days to consider)
seq_length = 20

# Create sequences for training
X_train, y_train = create_sequences(train_data, seq_length)
X_test, y_test = create_sequences(test_data, seq_length)
```

```python
data
```

```python
model = Sequential([
    SimpleRNN(units=50, input_shape=(seq_length, 1)),
    Dense(units=1)
])

# Compile the model
model.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error', metrics=['accuracy'])

```

```python
history = model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test), verbose=1)

```

```python
loss = model.evaluate(X_test, y_test, verbose=0)
print(f'Test loss: {loss}')

```

```python
predicted_prices = model.predict(X_test)

# Inverse transform the predicted prices to their original scale
predicted_prices = scaler.inverse_transform(predicted_prices)
```

```python
# Original test data (scaled)
original_prices = scaler.inverse_transform(y_test.reshape(-1, 1))

# Plotting
plt.figure(figsize=(14, 7))
plt.plot(original_prices, label='Actual Prices')
plt.plot(predicted_prices, label='Predicted Prices')
plt.title('Microsoft Stock Prices Prediction')
plt.xlabel('Time')
plt.ylabel('Stock Price')
plt.legend()
plt.show()
```
